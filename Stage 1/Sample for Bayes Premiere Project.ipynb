```python
#select the same columns for the main dataset
update_survey_data = survey_data[['Respondent', 'MainBranch', 'Hobbyist', 'Age', 'Age1stCode', 'ConvertedComp', 'Country',
                                 'DatabaseWorkedWith', 'DevType', 'EdLevel', 'Employment', 'Ethnicity', 'Gender', 'JobFactors',
                                 'LanguageWorkedWith', 'UndergradMajor', 'WebframeWorkedWith', 'WorkWeekHrs', 'YearsCode', 'YearsCodePro' ]]
update_survey_data
```

```python
#remove rows with missing values from data
survey_updated = update_survey_data.dropna()
survey_updated
```

```python
#summarise with pandas profiling
profile = pp.ProfileReport(survey_updated) 

profile.to_file("survey.html")
```

```python
#total countries that took the survey
country_data = survey_updated['Country'].value_counts()
country_data.to_frame()
```

```python
#total age, you can use this to remove the ages that don't make sense, ages less than 14 and greater than 60 should be removed
age_data = survey_data['Age'].value_counts()
age_data.to_frame()
```

```python
def age_label(row):
    age = row.Age
    if age <= 19:
        return "15 - 19"
    elif age <= 30:
        return "20 - 30"
    elif age <= 45:
        return "31 - 45"
    elif age <= 60:
        return "46 - 60"
    else:
        return "Above 60"
```

```python
survey_data["Age_label"] = survey_data.apply(age_label, axis=1)
```

### Demography of the Respondents

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
pie_data = survey_data.groupby("Age_label").Respondent.count()
pie_data = pie_data.to_frame()
```

```python
plt.pie(x='Respondent', 
        data=pie_data, 
        radius=1.5,
        autopct='%1.2f%%',
        normalize=True, 
       )
plt.legend(labels=['14 - 19', '20 - 30', '31 - 45', '46 - 60', 'Above 60'])
plt.show()
```

```python
race_count = survey_data.Ethnicity.str.split(';').explode().to_frame().reset_index().groupby('Ethnicity').count()
```

```python
race_count.rename({
    'index':'Count'
}, inplace=True, axis=1)
race_count
```

```python
plt.figure(figsize=(10,5))
plt.bar(x=race_count.index, height=race_count['Count'], log=True) 
# using log scale to cover for large difference between maximum data and the others
plt.xticks(rotation=90)
plt.show()
```
